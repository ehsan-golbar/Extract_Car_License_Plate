```python
# find dark window
import cv2
import numpy as np
from matplotlib import pyplot as plt
image_path = 'car.jpg'
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
if contours:
    largest_contour = max(contours, key=cv2.contourArea)
    x, y, w, h = cv2.boundingRect(largest_contour)
    dark_window = image[y:y+h, x:x+w]
    cropped_image_path = "darkWindowCropedFromCar.jpg"
    cv2.imwrite(cropped_image_path, dark_window)

else:
    print("Dark window overlay not found")
```

```python
# find white window
import cv2
import numpy as np
from matplotlib import pyplot as plt
cropped_image_path = "darkWindowCropedFromCar.jpg"
dark_window_image = cv2.imread(cropped_image_path)
gray_dark_window = cv2.cvtColor(dark_window_image, cv2.COLOR_BGR2GRAY)
_, thresh_white = cv2.threshold(gray_dark_window, 55, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(thresh_white, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
if contours:
    largest_contour = max(contours, key=cv2.contourArea)
    x, y, w, h = cv2.boundingRect(largest_contour)
    white_region = dark_window_image[y:y+h, x:x+w]
    white_region_path = "whiteWindowCropedFromCar.jpg"
    cv2.imwrite(white_region_path, white_region)
else:
    print("White region not found")
```

```python
# increase contrast
import numpy as np
import cv2
import json
from matplotlib import pyplot as plt
def read_this(image_file, gray_scale=False):
    image_src = cv2.imread(image_file)
    if gray_scale:
        image_src = cv2.cvtColor(image_src, cv2.COLOR_BGR2GRAY)
    else:
        image_src = cv2.cvtColor(image_src, cv2.COLOR_BGR2RGB)
    return image_src
def equalize_this(image_file, with_plot=False, gray_scale=False):
    image_src = read_this(image_file=image_file, gray_scale=gray_scale)

    r_image, g_image, b_image = cv2.split(image_src)

    r_image_eq = cv2.equalizeHist(r_image)
    g_image_eq = cv2.equalizeHist(g_image)
    b_image_eq = cv2.equalizeHist(b_image)

    image_eq = cv2.merge((r_image_eq, g_image_eq, b_image_eq))
    cmap_val = None

    cv2.imwrite("increasedContrastCar.jpg", image_eq)
    return image_eq
equalize_this("whiteWindowCropedFromCar.jpg", with_plot=False, gray_scale=False)

```
